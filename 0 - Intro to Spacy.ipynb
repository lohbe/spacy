```python
from __future__ import unicode_literals # used under python 2 only to prevent unicode errors
import spacy
from spacy import displacy
```

```python
nlp = spacy.load('en')
```

```python
text = open('austen-emma.txt').read().decode('utf8') # only use .decode() for python 2
```

```python
doc = nlp(text)
```

```python
for entity in doc.ents:
    print(entity.text, entity.label_)
```

```python
list(doc.sents)[2].text
```

```python
doc_dep = nlp(u'This is a sentence.')
#displacy.serve(doc_dep, style='dep')
displacy.serve(doc, style='ent')
```

```python
from wordcloud import WordCloud

#wordcloud = WordCloud().generate(text)

text2 = ""

for entity in doc.ents:
    if (entity.label_ == 'PERSON'):
        text2 += entity.text + " "
        
wordcloud = WordCloud().generate(text2)


import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

wordcloud = WordCloud(max_font_size=40).generate(text2)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
